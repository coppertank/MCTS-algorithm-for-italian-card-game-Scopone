```python
import itertools

def GreedyStrategy (legalMoves, table, deck, values):
    if isinstance(CheckForScopa(legalMoves, table), tuple):
        return CheckForScopa(legalMoves, table)

    else:
        possible_picks=AvoidScopa(legalMoves, table, deck)
        return BestMove(legalMoves, possible_picks, values)


def CheckForScopa(legalMoves, table):
    for card in legalMoves:
        if card[0] == sum(rank_table[0] for rank_table in table):
            return card # check if it's possible to do a scopa, the dictionary is sorted by values so it will pick with the card with highest value


def AvoidScopa(legalMoves, table, deck):
    list_table = [rank[0] for rank in table]
    list_deck = [rank[0] for rank in deck]
    possible_picks = []
    for card in legalMoves:
        if card[0] in list_table:
            copy = list_table[:]
            copy.remove(card[0])
            sum_table = sum(rank_table for rank_table in copy)
            if sum_table in list_deck:
                pass
            else:
                possible_picks.append(card)
        elif card[0] not in list_table:
            for r in range(2, len(list_table) + 1):
                for subset in itertools.combinations(list_table, r):
                    if sum(subset) == card[0]:
                        copy = list_table[:]
                        sum_table = sum(rank_table for rank_table in copy) - card[0]
                        if sum_table in list_deck:
                            pass
                        else:
                            possible_picks.append(card)
                    else:
                        copy = list_table[:]
                        copy.append(card[0])
                        sum_table = sum(rank_table for rank_table in copy)
                        if sum_table in list_deck:
                            pass
                        else:
                            possible_picks.append(card)
        else:
            pass
    return possible_picks


def BestMove(legalMoves, possible_picks, values):
    min_score = 0  # Inizializziamo il valore massimo con un valore molto piccolo
    max_score = 150
    best_choice = 0  # Inizializziamo la tupla massima come None

    if len(possible_picks) == 0:
        for card in legalMoves:
            if card in values:  # Verifichiamo se la tupla è una chiave nel dizionario
                value = values[card]  # Otteniamo il valore corrispondente alla tupla
                if value < max_score:  # Se il valore è minore del valore minimo attuale
                    max_score = value  # Aggiorniamo il valore minimo
                    best_choice = card  # Aggiorniamo la tupla minima

    else:
        for choice in possible_picks:
            if choice in values:  # Verifichiamo se la tupla è una chiave nel dizionario
                value = values[choice]  # Otteniamo il valore corrispondente alla tupla
                if value > min_score:  # Se il valore è maggiore del valore massimo attuale
                    min_score = value  # Aggiorniamo il valore massimo
                    best_choice = choice  # Aggiorniamo la tupla massima

    return best_choice
```
